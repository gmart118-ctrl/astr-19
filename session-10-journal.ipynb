```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# Use the normal (Gaussian) distribution
n = 1000
x = np.random.exponential(scale=1.0, size=n)

#create the histogram with 100 bins
plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
#label axes
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("expo-histo.pdf",bbox_inches="tight",facecolor="white")
```
