```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create an array x from 0 to 1 with 100 steps
x = np.linspace(0,1,100)
# Define a function that returns sin(x) and cos(x)
y = np.sin(x)
z = np.cos(x)
```

```python
# Create a multipanel plot (1 row, 2 columns)
f, axarr = plt.subplots(1, 2)

# Plot sin(x) vs. x in the left panel
axarr[0].plot(x,y)
axarr[0].set_xlabel('x')
axarr[0].set_ylabel('sin(x)')
axarr[0].set_title(r'$\sin(x)$')

# Plot cos(x) vs. x in the right panel
axarr[1].plot(x,z)
axarr[1].set_xlabel('x')
axarr[1].set_ylabel('cos(x)')
axarr[1].set_title(r'$\cos(x)$')

f.subplots_adjust(wspace=0.4)
plt.savefig("trig-graph.pdf",bbox_inches="tight",dpi=600,facecolor="white")
```
