```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
# Create 1000 random numbers uniformly distributed betweem 0 and 1
n = 1000
x = np.random.rand(n)

#create the histogram with 100 bins
plt.hist(x,bins=100,alpha=0.5,edgecolor="black")
#label axes
plt.ylabel("N per bin")
plt.xlabel("x")
plt.savefig("random-histo.pdf",bbox_inches="tight",facecolor="white")
```
